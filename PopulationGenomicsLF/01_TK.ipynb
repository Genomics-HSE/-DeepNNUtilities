**Приобразование генома вида 01 к TK**

На вход: name.npy - геномная последовательность, представленная в виде списка 0 и 1. 

В результате: Name.fa - та же геномная последовательность, но представленная в виде файла формата Fasta. 
 

```python
import numpy as np 
```

```python
x64_1 = np.load('trainings/64/x/1.npy')
```

```python
#генерируем список 
list_x64_1 = [ 'T' if x == 0 else 'K' for x in x64_1] 
list_x64_1
```

```python
lenght_x64_1 = len(list_x64_1)
lenght_x64_1
```
